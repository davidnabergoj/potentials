```python
from potentials.synthetic.funnel import GaussianExponentialPosterior
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
import torch

x = torch.linspace(-30, 30, 100)
y = torch.linspace(-20, 20, 100)
xx, yy = torch.meshgrid(x, y, indexing='xy')
xx_flat, yy_flat = xx.flatten(), yy.flatten()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(8, 2.5), sharey=True)

ax = axes[0]
torch.manual_seed(0)
u = GaussianExponentialPosterior(n_observations=1, likelihood_scale=0.1)
zz_flat = u(torch.concat([xx_flat[:, None], yy_flat[:, None]], dim=1))
zz = zz_flat.view_as(xx)
ax.contourf(xx, yy, zz, norm=colors.LogNorm())
ax.set_xlabel('Dimension 0')
ax.set_ylabel('Dimension 1')
ax.set_title(f'Sigma = {u.likelihood_scale:.1f}')

ax = axes[1]
torch.manual_seed(0)
u = GaussianExponentialPosterior(n_observations=1, likelihood_scale=1.0)
zz_flat = u(torch.concat([xx_flat[:, None], yy_flat[:, None]], dim=1))
zz = zz_flat.view_as(xx)
ax.contourf(xx, yy, zz, norm=colors.LogNorm())
ax.set_xlabel('Dimension 0')
# ax.set_ylabel('Dimension 1')
ax.set_title(f'Sigma = {u.likelihood_scale:.1f}')

ax = axes[2]
torch.manual_seed(0)
u = GaussianExponentialPosterior(n_observations=1, likelihood_scale=10.0)
zz_flat = u(torch.concat([xx_flat[:, None], yy_flat[:, None]], dim=1))
zz = zz_flat.view_as(xx)
ax.contourf(xx, yy, zz, norm=colors.LogNorm())
ax.set_xlabel('Dimension 0')
# ax.set_ylabel('Dimension 1')
ax.set_title(f'Sigma = {u.likelihood_scale:.1f}')

fig.tight_layout()
plt.show()
```
