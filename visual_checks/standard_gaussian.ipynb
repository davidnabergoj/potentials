```python
import torch
import matplotlib.pyplot as plt
from potentials.synthetic.gaussian.unit import StandardGaussian
```

```python
torch.manual_seed(0)

u = StandardGaussian(n_dim=5)
x = u.sample((100000,))

print(x.mean(dim=0))
print(x.var(dim=0))
print(u.mean)
print(u.variance)
print(u.second_moment)
```

```python
plt.figure()
plt.scatter(x[:10000, 0], x[:10000, 1], s=5, alpha=0.4)
plt.xlabel("Dim 0")
plt.ylabel("Dim 1")
plt.title("Sample scatterplot")
plt.tight_layout()
plt.show()
```
