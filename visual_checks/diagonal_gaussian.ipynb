```python
import torch
import matplotlib.pyplot as plt
from potentials.synthetic.gaussian.diagonal import (
    DiagonalGaussian0,
    DiagonalGaussian1,
    DiagonalGaussian2,
    DiagonalGaussian3,
    DiagonalGaussian4,
    DiagonalGaussian5
)
```

```python
# DG0

torch.manual_seed(0)

u = DiagonalGaussian0(n_dim=100)
x = u.sample((100000,))

plt.figure()
plt.scatter(x[:10000, 0], x[:10000, -1], s=5, alpha=0.4)
plt.xlabel("Dim 0")
plt.ylabel(f"Dim -1")
plt.title("Sample scatterplot")
plt.tight_layout()
plt.show()
```
