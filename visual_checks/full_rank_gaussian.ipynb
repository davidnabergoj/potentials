```python
import torch
import matplotlib.pyplot as plt
from potentials.synthetic.gaussian.full_rank import (
    FullRankGaussian0,
    FullRankGaussian1,
    FullRankGaussian2,
    FullRankGaussian3,
    FullRankGaussian4,
    FullRankGaussian5
)
```

```python
# DG0

torch.manual_seed(0)

u = FullRankGaussian0(n_dim=100)
x = u.sample((10000,))

plt.figure()
plt.scatter(x[:, 0], x[:, -1], s=5, alpha=0.4)
plt.xlabel("Dim 0")
plt.ylabel(f"Dim -1")
plt.title("Sample scatterplot")
plt.tight_layout()
plt.show()
```
